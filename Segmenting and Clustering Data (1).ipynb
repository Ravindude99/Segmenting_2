#### Importing the necessary packages

```python
from bs4 import BeautifulSoup 
import requests #Library for requests
import pandas as pd #Library for analysis of data
from geopy.geocoders import Nominatim #converting to latitude and longitude values
from IPython.display import Image #Library for displaying images
from IPython.core.display import HTML #Library for displaying images
from IPython.display import display_html #Library for displaying images
import numpy as np
print("Libraries Imported")
```

#### Obtaining the table by webscraping

```python
wiki_url = 'http://zims-en.kiwix.campusafrica.gos.orange.com/wikipedia_en_all_nopic/A/List_of_postal_codes_of_Canada:_M'
```

```python
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.title)
Toronto_Table = str(soup.table)
display_html(Toronto_Table,raw=True)
```

```python
df_Toronto_neighborhoods = pd.read_html(Toronto_Table)
df=df_Toronto_neighborhoods[0]
df.head()
```

#### Cleaning and Reprocessing the data

```python
df_1 = df[df.Borough != 'Not assigned']
df_2 = df_1.groupby(['Postcode', 'Borough'], sort = False).agg(', '.join)
df_2.reset_index(inplace = True)
df_2['Neighbourhood'] = np.where(df_2['Neighbourhood'] == 'Not assigned', df_2['Borough'], df_2['Neighbourhood'])
df_2
```

```python
df_2.shape
```

#### Obtaining the csv file for latitude and longitude

```python
import pandas
lat_lon_coords = pandas.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv')
lat_lon_coords.head()
```

#### Obtaining the table with latitude and longitude

```python
df_3 = pd.merge(df_2, lat_lon_coords, left_on = 'Postcode', right_on = 'Postal Code')
df_3.drop("Postal Code", axis=1, inplace=True) #Remove the column Postal Code
df_3.head()
```
